# Criando o modelo linear

```python
model = linear_model.LinearRegression()
model.fit(X,y)
```

```python
#B1
b1 = model.coef_
b1
```

```python
#B0
b0 = model.intercept_
b0
```

## Exibindo a fórmula da função linear

```python
print(b1[0])
```

```python
'y_pred = {} + {} * X0 + {} * X1 + {} * X2 + {} * X3 + {} * X4 + {} * X5 + {} * X6 + {} * X7 + {} * X8 + {} * X9'.format(model.intercept_, *list(model.coef_))
```

## Predizendo valores

```python
y_pred = model.predict(X)
y_pred
```

## Avaliando o nosso modelo linear

### Cálculo do scikit learn

```python
mse = metrics.mean_squared_error(y, y_pred)
print(mse)
```

### Cálculo usando python no braço

```python
mse_braco = sum((y - y_pred) ** 2) / len(y)
print(mse_braco)
```

# Criando modelo com a melhor feature

```python
model2 = linear_model.LinearRegression()
X_2 = [[i] for i in X[:,2]]
y_2 = [[i] for i in y]
model2.fit(X_2,y)
```

$ \beta_1 = \frac{ \sum_{i=1}^{n} (x_i - mean(x)) \times (y_i - mean(y))}{\sum_{i=1}^{n} (x_i - mean(x))^2} $

$ \beta_0 = mean(y) - \beta_1 \times mean(x) $

```python
b1_2 = np.sum((X_2 - np.mean(X_2)) * (y_2 - np.mean(y_2))) / np.sum((X_2 - np.mean(X_2)) ** 2)
print(b1_2)
```

```python
b0_2 = np.mean(y_2) - (b1_2 * np.mean(X_2))
print(b0_2)
```

```python
B1_2 = model2.coef_
B1_2
```

```python
B0_2 = model.intercept_
B0_2
```

```python
y_2_pred = model2.predict(X_2)
y_2_pred
```

```python
plt.grid()
plt.scatter(X_2,y_2)
plt.plot(X_2,y_2_pred, color="red")
```

```python
mse_2 = metrics.mean_squared_error(y_2, y_2_pred)
mse_2
```
